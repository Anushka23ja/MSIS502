```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=housing_df, x='housing_median_age', y='median_house_value')
plt.title('Relationship between Housing Age andhouse Values')
plt.xlabel('Housing Median Age')
plt.ylabel('Median House Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.scatterplot(data=housing_df, x='longitude', y='latitude', size='median_house_value', sizes=(20, 300), alpha=0.6, marker='+')
plt.title('Geographical Distribution of Houses with Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
