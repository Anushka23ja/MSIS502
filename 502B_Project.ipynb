```python
latest.columns
```

```python
fred_clean = latest[(latest['Original Debt-to-Income (DTI) Ratio'] < 100) & (latest['Original Interest Rate'] < 100) & (latest['Credit Score'] < 900 ) &(latest['Original Loan-to-Value (LTV)'] < 999)].dropna(subset=['Metropolitan Statistical Area (MSA) Or Metropolitan Division'])
fred_clean = fred_clean.drop(columns=fred_clean.loc[:, 'Pre-HARP Loan Sequence Number': 'Mortgage Insurance Cancellation Indicator'])
fred_clean = fred_clean.drop(columns=fred_clean.loc[:, 'Metropolitan Statistical Area (MSA) Or Metropolitan Division': 'Number of Units'])
fred_clean['Super Conforming Flag'] = fred_clean['Super Conforming Flag'].fillna('N')
fred_clean
```

```python
fred_clean.columns
```

```python


credit_score_bins = [0,580,670,740,800,850]
names = ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']
fred_clean['credit_score_bins'] = pd.cut(fred_clean['Credit Score'], bins=credit_score_bins, labels=names)

```

```python
fred_clean = pd.get_dummies(fred_clean, columns = ['Loan Purpose', 'credit_score_bins'], drop_first=True)
```

```python
fred_x = fred_clean.loc[:,['Loan Purpose_N', 'Loan Purpose_P','credit_score_bins_Excellent',
                            'credit_score_bins_Fair', 'credit_score_bins_Very Good',
                           'credit_score_bins_Good','Original Debt-to-Income (DTI) Ratio',
                           'Original Loan-to-Value (LTV)']]
fred_y = fred_clean.loc[:,'rate_spread']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

x_train, x_test, y_train, y_test = train_test_split(fred_x, fred_y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(model.coef_)
print(model.intercept_)
print(np.mean(np.abs(y_test - y_pred)))
print(r2_score(y_test, y_pred))
```
